```python
import pandas as pd
```

```python
df = pd.read_excel("0701_1000Microns.xlsx")
df
```

```python
ME_A_dis = df["ME_A_dis"][2:25]
print(ME_A_dis)
```

```python
fringe_A_dis = df["fringe_A_dis"][2:25]
print(fringe_A_dis)
```

```python
real_V = df["real_V"][2:25]
print(real_V)
```

```python
import matplotlib.pyplot as plt
plt.close("all")
%matplotlib notebook
```

```python
import numpy as np
```

```python
from scipy import stats
```

```python
x = fringe_A_dis
y = ME_A_dis
```

```python
info = stats.linregress(x,y)
print(info)
```

```python
xerror = np.full_like(x,0.6)
yerror = np.full_like(y,0)

plt.scatter(x,y)

plt.plot(x,x*info.slope+info.intercept,c = "green")

plt.plot(x,x,c = "red")

plt.errorbar(x,y,xerr=xerror,yerr=yerror,fmt = 'o',mfc = 'blue')

plt.xlabel('Acc. Displacement Calculated from Fringes (Microns)')
plt.ylabel('Acc. Displacement Measured by Capacitive Sensor')

t = '1000 $\mu m$ Capacitive Sensor Acc. Displacement vs. Fringe Count Acc. Displacement'
plt.text(-30, 85, t, ha='left', rotation=0, wrap=True)




plt.grid()
plt.show
```

```python
plt.close("all")
x = real_V
y = fringe_A_dis


plt.scatter(x,y)

plt.xlabel('Voltage (V)')
plt.ylabel('Acc. Displacement Calculated from Fringes (Microns)')

plt.title('Acc. Fringe Count Displacement vs. PZT Voltage')

plt.grid()
plt.show
```

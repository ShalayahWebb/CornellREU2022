```python
import pandas as pd
from matplotlib import pyplot as plt
from astropy.table import Table as Tb
%matplotlib notebook
```

```python
from scipy import stats
import numpy as np
```

```python
df1 = pd.read_excel("0630_1000Microns.xlsx")
df1 
```

```python
#data from run 1 of 1000 micron measurements 06/30
ME_A_dis1 = df1["ME_A_dis"][1:21]
print(ME_A_dis1)

fringe_A_dis1 = df1["fringe_A_dis"][1:21]
print(fringe_A_dis1)
```

```python
#cutting off data before jump in fringes
y1 = ME_A_dis1[:4]
x1 = fringe_A_dis1[:4]

#cutting off data after jump
y1_p = ME_A_dis1[4:]
x1_p =fringe_A_dis1[4:]


info1 = stats.linregress(x1,y1)
print(info1)

info1_2 = stats.linregress(x1_p,y1_p)
print(info1_2)
```

```python
df2 = pd.read_excel("0701_1000Microns.xlsx")
df2
```

```python
#data from run 2 of 1000 micron measurements 07/01
ME_A_dis2 = df2["ME_A_dis"][2:25]
print(ME_A_dis2)

fringe_A_dis2 = df2["fringe_A_dis"][2:25]
print(fringe_A_dis2)
```

```python
#cutting off at large fringe jump
y2 = ME_A_dis2[:6]
x2 = fringe_A_dis2[:6]

#cutting off data after large fringe jump
y2_p = ME_A_dis2[6:]
x2_p = fringe_A_dis2[6:]

info2 = stats.linregress(x2,y2)
print(info2)

info2_p = stats.linregress(x2_p,y2_p)
print(info2_p)
```

```python
print(x2_p)
print(y2_p)
```

```python
df3 = pd.read_excel("0706_1000Microns.xlsx")
df3
```

```python
#data from run 3 of 1000 micron measurements 07/06
ME_A_dis3 = df3["ME_A_dis"][2:23]
print(ME_A_dis3)

fringe_A_dis3 = df3["fringe_A_dis"][2:23]
print(fringe_A_dis3)
```

```python
y3 = ME_A_dis3[14:]
x3 = fringe_A_dis3[14:]

y3_p = ME_A_dis3[:14]
x3_p = fringe_A_dis3[:14]

info3 = stats.linregress(x3,y3)
print(info3)

info3_p = stats.linregress(x3_p,y3_p)
print(info3_p)
```

```python
print(y3)
print(x3)
```

```python

plt.close("all")

plt.scatter(x1, y1-info1.intercept, c = "green", linestyle=":")
plt.plot(x1,x1*info1.slope,c = "green", label = "Run 1 Slope")

plt.scatter(x1_p, y1_p-info1_2.intercept, c = "grey", linestyle=":")
plt.plot(x1_p,x1_p*info1_2.slope,c = "grey", label = "Run 1 Slope 2")


plt.scatter(x2, y2, c = "orange", linestyle = ":")
plt.plot(x2,x2*info2.slope,c = "orange", label = "Run 2 Slope")

plt.scatter(x2_p, y2_p-info2_p.intercept, c = "purple", linestyle = ":")
plt.plot(x2_p,x2_p*info2_p.slope,c = "purple", label = "Run 2 Slope 2")

plt.scatter(x3,y3-info3.intercept, c = "red",linestyle=":")
plt.plot(x3,x3*info3.slope,c = "red", label = "Run 3 Slope")

plt.scatter(x3_p,y3_p-info3_p.intercept, c = "pink",linestyle=":")
plt.plot(x3_p,x3_p*info3_p.slope,c = "pink", label = "Run 3 Slope 2")

plt.xlabel('Acc. Displacement Calculated from Fringes ($\mu m$)')
plt.ylabel('Acc. Displacement Measured by Capacitive Sensor ($\mu m$)')

t = '1000 $\mu m$ Capacitive Sensor Acc. Displacement vs. Fringe Count Acc. Displacement'
plt.text(-50, 80, t, ha='left', rotation=0, wrap=True)

plt.grid()
plt.legend()
plt.show()

plt.savefig('1000_microns_slopes_plot_oneSensor.pdf')
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from astropy.table import Table as Tb
%matplotlib notebook
```

```python
from scipy import stats
import numpy as np
```

```python
df = pd.read_excel("0802_500Microns.xlsx")
df
```

```python
#column of sensor 1 (channel 1)
ME_A_dis_CH1 = df["CH1 ME_A_dis"][3:28]
print(ME_A_dis_CH1)
```

```python
#column of sensor 2 (channel 2)
ME_A_dis_CH2 = df["CH2 ME_A_dis"][3:28]
print(ME_A_dis_CH2)
```

```python
x = ME_A_dis_CH2
y = ME_A_dis_CH1

info_Sensors = stats.linregress(x,y)
info_Sensors
```

```python
#plot of first sensor vs. second sensor
plt.close("all")

#plot of our reference data, channel 2
plt.plot(x,x,c = "green")


plt.scatter(x,y)


plt.xlabel('Cap. Sensor 2 Acc. Displacement ($\mu m$)')
plt.ylabel('Cap. Sensor 1 Acc. Displacement ($\mu m$)')
t = 'Run #1 500 ($\mu m$) 1st Capacitive Sensor Acc. Displacement vs. 2nd Capacitive Sensor Acc. Displacement'
plt.text(-16, 92, t, ha='left', rotation=0, wrap=True)


plt.grid()
plt.show

plt.savefig('Run 1 500 Microns Two Sensors.svg')
```

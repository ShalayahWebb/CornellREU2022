```python
import pandas as pd
from matplotlib import pyplot as plt
from astropy.table import Table as Tb
%matplotlib notebook
```

```python
from scipy import stats
import numpy as np
```

```python
df1 = pd.read_excel("0718_2000Microns1stMsmt.xlsx")
df1 
```

```python
#data from run 1 of 2000 micron measurements 07/18

ME_A_dis1 = df1["ME_A_dis"][2:27]
print(ME_A_dis1)

fringe_A_dis1 = df1["fringe_A_dis"][2:27]
print(fringe_A_dis1)
```

```python
y1 = ME_A_dis1[:14]
x1 = fringe_A_dis1[:14]

y1_p = ME_A_dis1[14:]
x1_p = fringe_A_dis1[14:]


info1 = stats.linregress(x1,y1)
print(info1)


info1_p = stats.linregress(x1_p,y1_p)
print(info1_p)

print(x1_p)
print(y1_p)

```

```python
slope1_2_2000um_std1 = np.std((1.0115776399825152,0.8891698617022448), dtype=np.float64)
print(slope1_2_2000um_std1)
```

```python
df2 = pd.read_excel("0718_2000Microns2ndMsmt.xlsx")
df2 
```

```python
#data from run 2 of 2000 micron measurements 07/18

ME_A_dis2 = df2["ME_A_dis"][2:27]
print(ME_A_dis2)

fringe_A_dis2 = df2["fringe_A_dis"][2:27]
print(fringe_A_dis2)
```

```python
y2 = ME_A_dis2[0:18]
x2 = fringe_A_dis2[0:18]

#cutting off where there is -19 for the fringe count
x2_p = fringe_A_dis2[18:]
y2_p = ME_A_dis2[18:]

info2 = stats.linregress(x2,y2)
print(info2)


info2_p = stats.linregress(x2_p,y2_p)
print(info2_p)

print(y2)
print(x2)
print(y2_p)
print(x2_p)
```

```python
df3 = pd.read_excel("0727_2000Microns_3rdMsmt.xlsx")
df3 
```

```python
#data from run 3 of 2000 micron measurements 07/27

ME_A_dis3 = df3["ME_A_dis_CH1"][3:28]
print(ME_A_dis3)

fringe_A_dis3 = df3["fringe_A_dis"][3:28]
print(fringe_A_dis3)
```

```python
#data before it reaches NaN for fringe acc. displacement
y3 = ME_A_dis3[0:8]
x3 = fringe_A_dis3[0:8]


#cutting off where there is NaN for the fringe displacement
x3_p2 = fringe_A_dis3[8:]
y3_p2 = ME_A_dis3[8:]

#linear regression for data before NaN
info3 = stats.linregress(x3,y3)
print(info3)


#linear regression for data after NaN fringe
info3_2 = stats.linregress(x3_p2,y3_p2)
print(info3_2)

```

```python

plt.close("all")

plt.scatter(x1, y1-info1.intercept, c = "green", linestyle=":")
plt.plot(x1,x1*info1.slope,c = "green", label = "Run 1 Slope")

plt.scatter(x1_p, y1_p-info1_p.intercept, c = "grey", linestyle=":")
plt.plot(x1_p,x1_p*info1_p.slope,c = "grey", label = "Run 1 Slope 2")


plt.scatter(x2, y2-info2.intercept, c = "orange", linestyle = ":")
plt.plot(x2,x2*info2.slope,c = "orange", label = "Run 2 Slope")

plt.scatter(x2_p, y2_p-info2_p.intercept, c = "purple", linestyle = ":")
plt.plot(x2_p,x2_p*info2_p.slope,c = "purple", label = "Run 2 Slope 2")

plt.scatter(x3,y3-info3.intercept, c = "red",linestyle=":")
plt.plot(x3,x3*info3.slope,c = "red", label = "Run 3 Slope")

plt.scatter(x3_p2,y3_p2-info3_2.intercept, c = "pink",linestyle=":")
plt.plot(x3_p2,x3_p2*info3_2.slope,c = "pink", label = "Run 3 Slope 2")


plt.scatter(x4,y4-info4.intercept, c = "blue",linestyle=":")
plt.plot(x4,x4*info4.slope,c = "blue", label = "Run 4 Slope")

plt.scatter(x4_p,y4_p-info4_p.intercept, c = "yellow",linestyle=":")
plt.plot(x4_p,x4_p*info4_p.slope,c = "yellow", label = "Run 4 Slope 2")

plt.scatter(x5,y5-info5.intercept, c = "cyan",linestyle=":")
plt.plot(x5,x5*info5.slope,c = "cyan", label = "Run 5 Slope")

plt.xlabel('Acc. Displacement Calculated from Fringes ($\mu m$)')
plt.ylabel('Acc. Displacement Measured by Capacitive Sensor ($\mu m$)')

t = '2000 $\mu m$ Capacitive Sensor Acc. Displacement vs. Fringe Count Acc. Displacement'
plt.text(-36, 95, t, ha='left', rotation=0, wrap=True)

plt.grid()
plt.legend()
plt.show()

plt.savefig('2000_microns_slopes_plot_oneSensor.pdf')
```
